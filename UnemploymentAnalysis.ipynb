# Sort Date in descending order

```python
data1['Date'] = pd.to_datetime(data1['Date'])
data1 = data1.sort_values(by='Date', ascending=False)
```

```python
data1.head()
```

```python
mean_unemployment_over_time = data1.groupby('Date')['Estimated Unemployment Rate (%)'].mean()
mean_unemployment_over_time
```

```python
plt.figure(figsize=(10, 6))
plt.plot(mean_unemployment_over_time.index, mean_unemployment_over_time.values, marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate Over Time')
plt.grid(True)

plt.show()
```

This line chart shows us the sharp rise in umemployment rate in the Covid-19 period

```python
fig = px.bar(data1, x = 'Region', y = "Estimated Unemployment Rate (%)", color = "Region", title = "Average unemploment Rate")
fig.update_layout(xaxis = {'categoryorder':'total descending'})
fig.show()
```

```python
data1["Area"].value_counts()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Area', y='Estimated Unemployment Rate (%)', data=data1)
plt.xlabel('Area')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate by Area')
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Area', y='Estimated Employed', data=data1)
plt.xlabel('Area')
plt.ylabel('Estimated Employed')
plt.title('Employed Proprtion by Area')
```

This shows that unemployment is much higher in the urban areas than in the rural areas in India
